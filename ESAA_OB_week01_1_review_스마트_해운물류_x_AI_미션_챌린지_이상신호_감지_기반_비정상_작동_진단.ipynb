# 1. 코드 흐름 요약

이 코드는 해운 물류 장비에서 발생하는 센서 데이터를 기반으로 **비정상 작동 여부를 분류하는 모델**을 구축하는 과정으로 구성되어 있다.

전체 흐름은 다음과 같다.

### ① 데이터 로딩 및 전처리

* 학습 데이터(train)와 테스트 데이터(test)를 불러온다.
* 결측치 처리, 범주형 변수 인코딩, 스케일링 등을 수행한다.
* 모델 학습에 필요한 입력 feature와 label을 분리한다.

### ② 모델 학습

* 여러 모델을 사용하여 각각 학습한다.
* 주요 모델

  * **TabNet** : 테이블 데이터에서 feature 선택을 자동으로 수행하는 딥러닝 모델
  * **MLP (Multi Layer Perceptron)** : 기본적인 신경망 기반 분류 모델
  * **AutoGluon** : 다양한 모델을 자동으로 학습하고 앙상블하는 AutoML 프레임워크

각 모델이 동일한 데이터로 학습되고 **예측 확률을 생성한다.**

### ③ 앙상블 (Ensemble)

* 각 모델의 예측 확률을 평균하거나 결합하여 **최종 예측 확률을 생성한다.**
* 여러 모델을 결합함으로써 단일 모델보다 **일반화 성능을 높인다.**

### ④ Hungarian Algorithm 적용

* 예측 결과를 최적의 매칭 형태로 조정하기 위해
  **Hungarian Algorithm(최적 할당 알고리즘)**을 사용한다.
* 이를 통해 예측 결과와 실제 label 구조를 더 잘 맞추도록 후처리한다.

### ⑤ 제출 파일 생성

* 테스트 데이터에 대한 최종 예측 결과를 생성한다.
* 제출 형식에 맞게 csv 파일로 저장한다.

# 2. 새롭게 알게 된 내용

### ① TabNet 모델

TabNet은 **테이블 데이터(Tabular Data)를 위한 딥러닝 모델**로,
feature selection을 attention mechanism으로 수행하는 특징이 있다.

즉,

* 트리 기반 모델(LightGBM 등) 대신
* **딥러닝 기반 tabular 모델**을 사용할 수 있다는 점을 알게 되었다.

---

### ② AutoGluon

AutoGluon은 AutoML 라이브러리로

* 여러 모델을 자동 학습
* 하이퍼파라미터 튜닝
* 자동 앙상블

을 수행해주는 프레임워크이다.

따라서 **모델 성능을 빠르게 높일 수 있는 방법**이라는 것을 알게 되었다.

---

### ③ Hungarian Algorithm 활용

Hungarian Algorithm은

* **최적 할당 문제 (assignment problem)** 를 해결하는 알고리즘이다.

예를 들어

* 예측 클래스
* 실제 클래스

간 매칭을 **최적의 조합으로 맞추는 데 사용**된다.

이 방법이 **머신러닝 후처리(Post-processing)** 에 사용될 수 있다는 점이 새로웠다.
