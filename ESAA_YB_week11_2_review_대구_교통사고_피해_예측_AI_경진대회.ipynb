# 🔹 대구 교통사고 피해 예측 AI 경진대회

https://dacon.io/competitions/official/236193/overview/description

### **수상작 리뷰 (요약 버전)**

---

## **1. 주제 및 데이터 요약**

### **주제**

* 대구광역시 교통사고 정보를 기반으로 **피해 정도(ECLO)**를 예측하는 회귀 문제
* ECLO = 사망·중상·경상·부상자 수의 가중합
* **평가지표: RMSLE**

### **사용 데이터**

* **대회 제공 데이터**: 사고일시·요일·구/동·차종·사상자 수 등 사고 정보
* **외부 데이터 대거 활용**

  * 보안등, CCTV, 어린이 보호구역, 주차장 데이터
  * 2007–2018년 과거 교통사고 GIS (공공데이터 포털)
* train/test와 **동 단위로 매핑 → 환경 기반 위험도 반영**

---

## **2. EDA 핵심 인사이트**

* **시간대**: 새벽(0~5시) 사고의 ECLO가 가장 높음
* **요일**: 평일 < 토요일 < **일요일**
* **지역**: 달성군·동구 등 외곽 지역에서 피해도가 높음
* **CCTV 제한속도**가 높을수록 사고 피해가 큼
* 사고 피해는 **지역·시간·도로환경 패턴이 뚜렷**함

➡ 이런 패턴들을 이후 **파생변수(feature)** 로 전환한 점이 강점

---

## **3. 전처리 및 Feature Engineering 요약**

### **핵심 피처 설계**

* **시간대 그룹화** (새벽/오전/오후/야간)
* **요일 그룹화** (평일 / 토요일 / 일요일)
* **주기적(cyclical) 인코딩**

  * `sin/cos` 변환하여 시간·날짜의 주기성 반영
* **구/동별 위험도 더미 변수**

  * 과거 평균 ECLO≥5 지역 표시 → 지역 latent risk 반영
* **동별 누적 사상자 수 집계**
* **외부 데이터 기반 환경 피처**

  * CCTV 수·제한속도
  * 보안등·어린이 보호구역 수
  * 주차장 규모 등

➡ “환경+기반시설+시공간”을 모두 반영한 **풍부한 피처 엔지니어링**이 성능 핵심

---

## **4. 모델링 전략**

* 모델: **CatBoost + XGBoost**

  * CatBoost → 범주형·지역 정보 강함
  * XGBoost → 연속형·파생변수 학습에 강함
* 검증: **Multi-StratifiedKFold**

  * ECLO의 불균형을 고려한 층화 K-fold
* 여러 **seed 앙상블**로 성능 안정화
* 과도한 하이퍼파라미터 튜닝보다
  → **피처 품질 + 검증 안정성**에 집중한 전략

---

## **5. 차별점 & 배울 점 (핵심만)**

### 🔹 **차별점**

1. **외부 데이터 적극 수집 + 동 단위 매핑**
   → 도시 인프라 기반 위험도 반영
2. **지역 위험도 더미 + 시간/요일 주기성 처리**
   → 사고 패턴을 구조적으로 이해
3. **데이터 중심 접근**

   * 튜닝보다 **Feature Engineering**과 **검증 전략**이 중심
4. **실패한 시도도 문서화**

   * gpkg 매핑 실패, ECLO 반올림 실패 등
     → 분석 과정의 투명성·학습 가치 높음

### 🔹 **배울 점**

* 단순 사고 데이터만 보지 말고 **환경·시설 데이터까지 확장할 것**
* 지역 단위 위험도를 수치화하는 **집계 기반 파생변수**가 강력
* 범주형 많을 때 CatBoost, 일반 수치형 많을 때 XGBoost 조합 유효
* 검증(Validation)이 모델성능의 절반
