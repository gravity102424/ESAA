# 상추의 생육 환경 생성 AI 경진대회


https://dacon.io/competitions/official/236033/overview/description

## 수장작 리뷰

---

## 1. **주제 & 데이터**

* **목표:**

  * 상추의 **일별 잎중량(leaf weight)**을 예측하는 모델 개발
  * 예측모델 기반으로 **생육환경(시간별 환경변수) 생성/설계 모델** 제안
* **데이터 구성:**

  * 케이스별(28개) **시간당 환경 데이터** (DAT별 0~27일, 1시간 간격)
  * 케이스별 **일별 잎중량(1~28일)**
  * **테스트 데이터도 동일 구조, 잎중량 숨김 상태**
* **환경 변수에 허용 범위 존재** → 현실적인 환경 제약 고려 필요

---

## 2. **코드 리뷰 – 핵심 흐름**

공유 코드/후기 기반으로 대부분의 상위 코드가 다음 구조를 가짐:

1. **환경 데이터(시계열) 전처리**

* 1시간 단위 환경값 → 일 단위 요약(평균, 합, 변화량 등)
* 케이스별 시퀀스 구조 유지
* 스케일링 및 이상치 처리 중요

2. **예측 모델링**

* Baseline: **LSTM/GRU** 중심의 시계열 회귀
* 또는 환경 요약 후 **LightGBM/XGBoost/MLP** 사용
* 케이스 기반 CV로 과적합 방지

3. **생육환경 생성(최적화)**

* 예측모델을 반복 호출하며
  → “환경 변수 조절 → 예상 잎중량 계산 → 최적 조합 탐색”
* 제한범위 안에서 가장 성능 높은 환경 시나리오를 생성
* 이 부분이 대회 순위에서 큰 차이를 만드는 핵심

4. **정성 평가 제출**

* 생성된 환경이 “왜 좋은지”
* 생육 이해 기반의 설명, 시각화 → 높은 점수

---

## 3. **차별점**

* 단순 예측 대회가 아니라 **“예측 + 환경 생성(최적화)” 두 가지를 모두 수행**
* 고해상도 **시계열 환경 데이터(1시간 단위)**
* 최종 평가 70%가 **정성 평가** → 설명과 인사이트 매우 중요

---

## 4. **배울 점**

* 시계열 회귀 모델 설계(LSTM, MLP, LightGBM 등)
* 환경 변수 엔지니어링(일단위 요약·변화율·패턴)
* 모델 기반 **시나리오 생성/최적화** 접근법
* CV 설계와 재현성 높은 코드 관리
* 최적화 제안서를 **정성적으로 잘 설명하는 능력**
